```python

def change_sentece(charOffset,sent,char_drug):
    flag = False
    posCollect = charOffset.split(';')
    anssent = sent
    for pos in posCollect:
        drug = ""
        pos_begin,pos_end = pos.split('-')
        pos_begin = int(pos_begin) 
        pos_end = int(pos_end) 
        for i in range(pos_begin,pos_end + 1):drug+= char_drug
        for i in range(pos_begin - 1,-1,-1):
            
            if(anssent[i] == " "):break
            flag = True
            anssent =  anssent[0:i] + char_drug +anssent[i+1:]
        for i in range(pos_end + 1,len(anssent)):
            #print(i)
            if(anssent[i] == " "):break
            flag = True
            anssent =  anssent[0:i] + char_drug +anssent[i+1:]
        if flag:print(sent)
        anssent =  anssent[0:pos_begin] + drug +anssent[pos_end+1:]
    return anssent
```

```python
def A_to_drug0(word,charr):
    if(len(word) == 0):return word
    for alpha in word:
        if alpha!= charr:
            return word
    return "drug" + charr
    
```

```python
def nltk_and_drugA(sentence):
    
    nltk_text = word_tokenize(sentence)
    words = nltk_text
    #words = [word for word in nltk_text if word.isalpha()]
    after_replace = []
    for word in words:
        word = A_to_drug0(word,'A')
        word = A_to_drug0(word,'B')
        word = A_to_drug0(word,'N')
        after_replace.append(word)
        #print(word)
    #word_lower = [word.lower() for word in after_replace]
    #word_no_stop = [word for word in word_lower if word not in stopwords.words('english')]
    
    #return " ".join(word_lower)
    #return " ".join(word_no_stop)  
    return " ".join(after_replace)
```

```python
def get_nltk(sentence):
    
    nltk_text = word_tokenize(sentence)
    word_lower = [word.lower() for word in nltk_text]
    
    #return " ".join(word_lower)
    return " ".join(word_lower)    
```

```python
def create_dict(root,idd):
    fout = open("test7.txt","a")
    #fout = open("origin.txt","a")
    for sentence in (root):
        entity_pos = {}
        entity_ids = []
        mysentece = sentence.attrib["text"]
        for entity in sentence:
            if entity.tag == "entity":
                entity_pos[entity.attrib["id"]] = entity.attrib["charOffset"]
                entity_ids.append(entity.attrib["id"])
        for pair in sentence:
            if pair.tag == "pair":
                new_sent = mysentece
                for entity_id in entity_ids:
                    if entity_id == pair.attrib["e1"] :
                        new_sent =change_sentece(entity_pos[entity_id],new_sent,"A")
                    elif entity_id == pair.attrib["e2"]:
                        new_sent =change_sentece(entity_pos[entity_id],new_sent,"B") 
                    else:
                        new_sent =change_sentece(entity_pos[entity_id],new_sent,"N")
                new_sent = nltk_and_drugA(new_sent)
                typee = pair.attrib["ddi"]
                if pair.get("type"):
                    typee = pair.attrib["type"]
               # print(new_sent)
               # print(typee)
                #fout.write(str(idd)+"\n")
                fout.write(new_sent + "\n")
                fout.write(typee +"\n")
                idd += 1
    return idd
                
                
```

```python
#file_path = "Train/MedLine/"
file_path = "Test/Test_data_DDI/MedLine"
files = os.listdir(file_path)
idd = 5265
for file in files:
    path = os.path.join(file_path,file)
    tree = read_xml(path)
    idd = create_dict(tree.getroot(),idd)
```
